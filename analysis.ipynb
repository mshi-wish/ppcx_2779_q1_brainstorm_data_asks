```python
from datetime import datetime, timezone
import pytz

country_timezone = {'US': 'America/Los_Angeles',\
                    'MX': 'America/Mexico_City',\
                    'IT': 'Europe/Rome',\
                    'FR': 'Europe/Paris',\
                    'ES': 'Europe/Madrid',\
                    'GB': 'Europe/London',\
                    'DE': 'Europe/Berlin',\
                    'BR': 'America/Sao_Paulo',\
                    'AU': 'Australia/Sydney',\
                    'CA': 'America/Toronto',\
                    'CL': 'America/Santiago',\
                    'PT': 'Europe/Lisbon',\
                    'HU': 'Europe/Budapest',\
                    'NL': 'Europe/Amsterdam',\
                    'CH': 'Europe/Zurich',\
                    }

def unix_time_to_local_date(date, country):
    if pd.isna(date):
        return np.nan
    if country in country_timezone:
        local_timezone = country_timezone[country]
        return datetime.fromtimestamp(date, pytz.timezone(local_timezone)).date()#.strftime("%Y-%m")
    else:
        return datetime.fromtimestamp(date, pytz.timezone('UTC')).date()#.strftime("%Y-%m")
```

```python
data['picked_up_date_local'] = data.apply(\
    lambda x: unix_time_to_local_date(date=x['picked_up_time'], country=x['store_country']), axis=1)

data['picked_up_dow_local'] = data['picked_up_date_local'].apply(\
                    lambda x: np.nan if pd.isna(x) else x.weekday())
```

# 9. Are these users first time Wish buyers or repeat buyers?
Have they picked up a local order before?
Have they missed an order pick up before?


```python
# Have they picked up a local order before?
def new_fusion_buyer(x):
    '''
    if by the txn time, the user does not have any Fusion order scanned by store, 
    then define s/he as a new fusion buyer
    '''
    
    if x['transaction_time'] < x['first_fusion_scan_time']:
        return 'new fusion buyer'
    else:
        return 'repeat fusion buyer'
    
data['is_new_fusion_buyer'] = data.apply(new_fusion_buyer, axis=1)
```

```python
data.groupby(['is_new_fusion_buyer']).agg(
    {'is_picked_up_14d':'mean',
    'm_transaction_id':'count',
    'user_id':'nunique'}
).to_clipboard()
```

```python
data.groupby(['store_country','is_new_fusion_buyer']).agg(
    {'is_picked_up_14d':'mean',
    'm_transaction_id':'count',
    'user_id':'nunique'}
).to_clipboard()
```

```python
(data.groupby(['store_country','is_new_fusion_buyer']).agg(
    {
    'm_transaction_id':'count',
    'user_id':'nunique'}
) / data.groupby(['store_country']).agg(
    {
    'm_transaction_id':'count',
    'user_id':'nunique'})).to_clipboard()
```

```python
# Have they missed an order pick up before?

def missed_pickup_buyer(x):
    '''
    if txn time > first_missed_fusion_pickup_scan_time (the last day of the pickup window of the missed) ,
    then the user has missed a pickup before
    '''
    if x['transaction_time'] > x['first_missed_fusion_pickup_scan_time']:
        return 'has missed pickup before'
    else:
        return 'never missed pickup or new buyer'

data['is_missed_pickup_buyer'] = data.apply(missed_pickup_buyer, axis=1)

```

```python
data.groupby(['is_missed_pickup_buyer']).agg(
    {'is_picked_up_14d':'mean',
    'm_transaction_id':'count',
    'user_id':'nunique'}
).to_clipboard()
```

```python
data.groupby(['store_country', 'is_missed_pickup_buyer']).agg(
    {'is_picked_up_14d':'mean',
    'm_transaction_id':'count',
    'user_id':'nunique'}
).to_clipboard()
```

```python
(data.groupby(['store_country','is_missed_pickup_buyer']).agg(
    {
    'm_transaction_id':'count',
    'user_id':'nunique'}
) / data.groupby(['store_country']).agg(
    {
    'm_transaction_id':'count',
    'user_id':'nunique'})).to_clipboard()
```
